##  Compare genomes of closley related species  

Two relatively closely related pathogenic fungi, _Fusarium oxysporum_ (_Fo_) and _Fusarium graminearum_ (_Fg_) have a very different chromosome number. _Fg_ has 4 chromosomes, while _Fo_ has 15. They also differ in size: _Fg_ has a genome of 36.36 Mb (megabases: 1 million A,T,C or G) and _Fo_ has 57.91 Mb. We want to compare these two genomes and see whether for example one chromsoome in _Fg_ corresponds to two or more chromosomes in _Fo_.

For this we need to download and align the genomes of _Fo_ and _Fg_. We will use Circos to visualize these alignments.

### 2. Align the genomes of _Fusarium oxysporum_ and _Fusarium graminearum_  

Next, we want to compare these genomes, to see which genomic regions are homologous to eachother. For this, we will use [MUMmer](http://mummer.sourceforge.net/). MUMmer is a package that contains programs to align genomes, filter aligned regions and plot algnments in dotplots. MUMmer is well-documented.

We will use promer to align _Fo_ and _Fg_. Promer first translates dna sequences into protein and then aligns those. It is more sensitive over longer evolutionary distances than nucmer (that aligns DNA sequences directly).  

I installed MUMmer version 3.23 in my /Applications folder so I will run the genome alignment like this:

### 3.1 Karyotype file
The karyotype file also contain information on the colors of the chromosomes:  
>chr - ID LABEL START END COLOR

So for example, chr 1 in _Fg_ is `CM000574.1`, and it has 11697295 basepairs and if we want plot these in orange we write:  
`chr - CM000574.1 1 0 11697295 orange`  
  
Let's create a such a file using Python. We will plot the chromosomes of different species in different colors in order to separate them. 




```python
# first karyotype file
import glob

# I run this notebook from the working directory hence I don't need to
# specify that path; instead of path/to/CircosPlot_FgFo/data/genomes/
# I can just say data/genomes/
indir   = 'data/genomes/' 
outdir  = 'circos/input/karyotypes/'
def karyotype_file_from_fasta(fasta_fname, out_fname, color = 'grey'):
    '''
    This function takes a fastafile (fasta_fname) and "translates" this 
    to a karyotype file (out_fname). Grey is the default color for chromosomes.
    '''
    outfile = open(out_fname, 'w')
    l      = 0 # counter for the sequence length
    cindex = 1 # counter for the chromosomes, note that this only works 
               # because the fastafile is ordered!
    cid    = ''
    for line in open(fasta_fname).readlines():
        if line[0] == '>': 
            '''
            This is the fastaheader. We need to store the name in this header 
            (we assign it to the variable cid), while we count the length of 
            the subsequent lines (the sequence of this chromosome), until we 
            encounter the next header
            '''
            if l > 0: 
                '''
                If l > 0 it means it is not the first sequence.
                We first have to process the previous sequence before we can move on 
                with this one
                '''    
                #process previous sequence
                outfile.write('chr - '+cid+' '+str(cindex)+' 0 '+' '+str(l)+' '+color+'\n')
                #make everything ready for this one
                l = 0
                cindex += 1
                
            '''If you encounter a header, store it in cid.
            '''
            cid = line[1:].strip().split()[0]
            print(cid, cindex)
                
        else:
            ''' 
            If it is not header it is part of the sequence, add the length of this line. 
            str.strip() takes of any newlines, spaces etc. so we don't count those.
            '''
            l += len(line.strip())
        
    ''' 
    When you've reached the end of the file, you still need to process the last header. 
    We do this here.
    '''
    if l > 0:
        outfile.write('chr - '+cid+' '+str(cindex)+' 0 '+' '+str(l)+' '+color+'\n')
            
    outfile.close()
    
colors = ['green', 'orange']
filenames = glob.glob(indir+'*.fasta')

if len(colors) != len(filenames): #check if your list of colors is long enough 
    print("I need more colors, I have ", len(colors), " color(s) for ", len(filenames), " files")
else:  
    for i, fasta_fname in enumerate(glob.glob(indir+'*.fasta')):
        print(fasta_fname.split('/')[-1].split('.fasta')[0],colors[i])
        out_fname = outdir + fasta_fname.split('/')[-1].split('.fasta')[0]+'.txt'
        karyotype_file_from_fasta(fasta_fname, out_fname, color = colors[i])
    
```

### 3.3 Links
#### 3.3.1 Write links file from MUMmer coords file
We now have plotted the chromosomes of _Fo_ and _Fg_, but we want to see whther and how different chromosomes in these two species correspond to eachother. Circos is desiged to show links between ordered sets, so let's tart using it what is was designed to do.  
  
Link files have the [following format](http://circos.ca/documentation/tutorials/links/basic_links/lesson):  
>chrA from to chrB from to

We can get this information from the coords file:

```python
lines   = open('results/genome_alignments/Fol4287__vs__Fg_PH1.promer_def.coords').readlines()[5:]
print(lines[0])
columns = lines[0].split()
print(columns)

```

Now let's write the 'translation' from this format to the circos link format to a function

```python
def coords_to_links(coords_fname, links_fname):
    linksfile = open(links_fname, 'w')
    lines     = open(coords_fname).readlines()[5:] #skip header
    for line in lines:
        columns = line.strip().split()
        
        if len(columns) != 17:
            print('Unexpected fileformat, can not continue!')
            break
            
        chrA  = columns[15]
        chrB  = columns[16]
        fromA = columns[0]
        toA   = columns[1]
        fromB = columns[3]
        toB   = columns[4]
        
        linksfile.write(chrA+' '+fromA+' '+toA+' '+chrB+' '+fromB+' '+toB+'\n')
    linksfile.close()
    print(coords_fname,": Done")
    
coords_fname ='results/genome_alignments/Fol4287__vs__Fg_PH1.promer_def.coords'
links_fname  = 'circos/input/links/links_Fo_vs_Fg.txt'
coords_to_links(coords_fname, links_fname)
               
```

We get an error that we're trying to display to many links. We can either reset the maximum number of links, or we can filter the links we plot. Let's do the latter:


```python
def coords_to_links(coords_fname, links_fname, min_length = 0):
    linksfile = open(links_fname, 'w')
    lines     = open(coords_fname).readlines()[5:] #skip header
    for line in lines:
        columns = line.strip().split()
        
        if len(columns) != 17:
            print('Unexpected fileformat, can not continue!')
            break
            
        chrA  = columns[15]
        chrB  = columns[16]
        fromA = columns[0]
        toA   = columns[1]
        fromB = columns[3]
        toB   = columns[4]
        
        lenA  = int(columns[6])
        lenB  = int(columns[7])
        if lenA >= min_length and lenB >= min_length: 
            linksfile.write(chrA+' '+fromA+' '+toA+' '+chrB+' '+fromB+' '+toB+'\n')
    linksfile.close()

coords_fname = 'results/genome_alignments/Fol4287__vs__Fg_PH1.promer_def.coords'
min_length   = 1000
links_fname  = 'circos/input/links/links_Fo_vs_Fg.min_length'+str(min_length)+'bp.txt'
coords_to_links(coords_fname, links_fname, min_length = min_length)
```

```python
import glob

def coords_to_links(coords_fname, links_fname, chr2color = None, min_length = 0):
    linksfile = open(links_fname, 'w')
    lines     = open(coords_fname).readlines()[5:] #skip header
    for line in lines:
        columns = line.strip().split()
        
        if len(columns) != 17:
            print('Unexpected fileformat, can not continue!')
            break
            
        chrA  = columns[15]
        chrB  = columns[16]
        fromA = columns[0]
        toA   = columns[1]
        fromB = columns[3]
        toB   = columns[4]
        
        lenA  = int(columns[6])
        lenB  = int(columns[7])
        if lenA >= min_length and lenB >= min_length: 
            out = chrA+' '+fromA+' '+toA+' '+chrB+' '+fromB+' '+toB
            
            if chr2color != None:
                if chrA in chr2color:
                    linksfile.write(out+' color='+chr2color[chrA]+'\n')
                elif chrB in chr2color:
                    linksfile.write(out+' color='+chr2color[chrB]+'\n')
            else:
                linksfile.write(out+'\n')
    linksfile.close()


def karyotype_file_from_fasta(fasta_fname, out_fname, chr2color = None):

    outfile = open(out_fname, 'w')
    l      = 0 #counter for the sequence length
    cindex = 1 #counter for the chromosomes
    cid    = ''
    color  = 'grey'
    for line in open(fasta_fname).readlines():
        if line[0] == '>':
            if l > 0:
                outfile.write('chr - '+cid+' '+str(cindex)+' 0 '+' '+str(l)+' '+color+'\n')
                l = 0
                cindex += 1
            cid = line[1:].strip().split()[0]
            if chr2color != None and cid in chr2color:
                color = chr2color[cid]
            print(cid, cindex)
                
        else:
            l += len(line.strip())
        
    #last one:
    if l > 0:
        outfile.write('chr - '+cid+' '+str(cindex)+' 0 '+' '+str(l)+' '+color+'\n')
            
    outfile.close()
    
# write karyotype files from fasta
chr2color = {}
chr2color['CM000574.1'] = 'blue'
chr2color['CM000575.1'] = 'yellow'
chr2color['CM000576.1'] = 'green'
chr2color['CM000577.1'] = 'red'

for i, fasta_fname in enumerate(glob.glob('data/genomes/*.fasta')):
    
    out_fname = 'circos/input/karyotypes/' + fasta_fname.split('/')[-1].split('.fasta')[0]+'.color-coded.txt'
    karyotype_file_from_fasta(fasta_fname, out_fname, chr2color = chr2color)

coords_fname = 'results/genome_alignments/Fol4287__vs__Fg_PH1.promer_def.coords'
links_fname  = 'circos/input/links/links_Fo_vs_Fg.min_length1000bp.color-coded.txt'
coords_to_links(coords_fname, links_fname, chr2color = chr2color, min_length = 1000)
```

We can see that chromosom1 in _Fg_ corresponds roughly to chr 7, half of chr 4 and chr 1 in _Fo_, but e.g. for chr 02 in _Fg_ it is tricky to see which part belongs to which chromosome in _Fo_. We can further improve our visualization by color-coding the chromosomes in _Fo_, rather than the chromosomes in _Fg_ and make 4 plots: one per chromosome.

Circos implements [brewer palettes for qualitative data] (http://colorbrewer2.org/#type=qualitative&scheme=Set3&n=12) for up to 12 categories. _Fo_ has 15 chromosomes, but chromosome 3, 6, 14 and 15 have very little regions that are homologous to regions in _Fg_ (are lineage-specific), hence we use a single color (e.g. black) for those. 

If we use `set3` of 11 colors (15 - 4 lineage-specific chromosomes) corresponding to categories (rather e.g. sequential data):`qual`, we can refer to them with `set3-11-qual-N`, where N is the number 1 to 11. 

```python
import glob, os

# Write karyotype files from fasta
chr2color = {}
chr2color['CM000589.1'] = 'set3-11-qual-1'
chr2color['CM000590.1'] = 'set3-11-qual-2'
chr2color['CM000591.1'] = 'light-grey' #chr 3: lineage-specific
chr2color['CM000592.1'] = 'set3-11-qual-3'
chr2color['CM000593.1'] = 'set3-11-qual-4'
chr2color['CM000594.1'] = 'light-grey' #chr 6: lineage-specific
chr2color['CM000595.1'] = 'set3-11-qual-5'
chr2color['CM000596.1'] = 'set3-11-qual-6'
chr2color['CM000597.1'] = 'set3-11-qual-7'
chr2color['CM000598.1'] = 'set3-11-qual-8'
chr2color['CM000599.1'] = 'set3-11-qual-9'
chr2color['CM000600.1'] = 'set3-11-qual-10'
chr2color['CM000601.1'] = 'set3-11-qual-11'
chr2color['CM000602.2'] = 'light-grey' #chr 14: lineage-specific
chr2color['CM000603.1'] = 'light-grey' #chr 15: lineage-specific


# Write the karyotype file for Fol4287
Fol_karyo_fname = 'circos/input/karyotypes/Fol4287_chr_color-coded.txt'
fasta_fname     = 'data/genomes/Fol4287.fasta'
karyotype_file_from_fasta(fasta_fname, Fol_karyo_fname, chr2color = chr2color)

#write the links file, now color them according to the Fol4287 chromosomes instead of the Fg_PH1 chromosomes
coords_fname = 'results/genome_alignments/Fol4287__vs__Fg_PH1.promer_def.coords'
links_fname  = 'circos/input/links/links_Fo_vs_Fg.min_length1000bp.color-coded_Fol.txt'
coords_to_links(coords_fname, links_fname, chr2color = chr2color, min_length = 1000)

# We need to make 4 plots, one for each Fg chromosome.
# There are a lot of different ways to do this. 
#
# I decided here to write a separate file for each chromosome in F. graminearum, 
# splitting up the existing karyotype file.
#
# Moreover, as I am looping over the chromosomes in the karyotypefile, 
# I also write the config file.
#
# First open a 'template' config file and store it's lines, so we can write them to 
# the 4 new chromosome-specific config files:

config_lines = open('circos/input/config_files/Fo_vs_Fg.circos.incl_colored_links.filtered.conf').readlines()

'''
You can run the code below, to print the line number and line of the config file to make sure 
you're changing the correct line:

for i,line in enumerate(config_lines):
    print(i,line.strip())
    
'''

# loop over the Fg chromosomes:
karyotype_file_Fg = 'circos/input/karyotypes/Fg_PH1.txt'
for i, line in enumerate(open(karyotype_file_Fg).readlines()):
    
    # write the karyotype file
    chrom_spec_fname = 'circos/input/karyotypes/Fg_chr'+str(i+1)+'.txt'
    Fg_chrom_file = open(chrom_spec_fname, 'w')
    Fg_chrom_file.write(line.replace('orange', 'grey')) # change color of the chromosomes 
                                                        # (one of the Fo chrs is orange)
    Fg_chrom_file.close()
    
    # Make the config file:
    config_fname         = 'circos/input/config_files/Fo_vs_Fg_chr'+str(i+1)\
                                +'.circos.incl_colored_links.filtered.conf'
    # change the line with the name of the config file
    config_lines[0] = '# '+config_fname+'\n'
    # change the line listing the karyotype files, to use the chr-specific karyotype file for Fg
    config_lines[2] = 'karyotype = '+chrom_spec_fname+','+Fol_karyo_fname+'\n'
    # change the line with the file to use for the links
    config_lines[31] = 'file\t= '+links_fname+'\n'
    
    # Write the config file
    Fg_chrom_config_file = open(config_fname, 'w')
    for line in config_lines:
        Fg_chrom_config_file.write(line)
    Fg_chrom_config_file.close()
    
    # Finally, run circos using this config file:
    circos_command = '/Applications/circos-0.69-6/bin/circos' # path to Circos
    circos_command +=' -conf '+config_fname                   # the config file we want circos to use
    circos_command +=' -outputdir circos/output'              # where to save the output
    # filename of the generated plot
    circos_command +=' -outputfile circos_Fg_chr'+str(i+1)+'.vs.Fol4287_with_labels_and_colored_links.png' 

    print(circos_command)
    os.system(circos_command)


```
